```
# Turn all classes to 0
import os
from tqdm import tqdm
import numpy as np


mydir = "/content/ts/ts"

for file in tqdm(os.listdir(mydir), total=len(os.listdir(mydir))):
    if file.endswith(".txt"):
        f = open(f"{mydir}/{file}", "r")
        fileslines = f.read().splitlines()
        myTxtInfo = []
        for line in fileslines:
            info = line.rsplit(" ")
            x_center = info[1]
            y_center = info[2]
            width = info[3]
            height = info[4]
            myTxtInfo.append([str(0), x_center, y_center, width, height])
        # # remove the last \n
        # myTxtInfo = np.array(myTxtInfo)
        # myTxtInfo = np.delete(myTxtInfo,len(myTxtInfo)-1)
        
        # overwrite my txt files
        # f = open(f"{mydir}/{file}", "w")
        # f.write(myTxtInfo)
        # f.close()
        with open(f"{mydir}/{file}", "w") as txt_file:
            for line in myTxtInfo:
                txt_file.write(" ".join(line) + "\n") # works with any number of elements in a line
```
