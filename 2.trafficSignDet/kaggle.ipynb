```
import os
from tqdm import tqdm

mydir = "/content/ts/ts"
# take all txt files
txtFiles = []
for file in tqdm(os.listdir(mydir), total=len(os.listdir(mydir))):
    curVar = file.rsplit(".")
    curPath = os.path.join(mydir, file)
    if int(curVar[0]) < 800:
        if file.endswith(".txt"):
            !mv $curPath /content/traffic_signs/labels/train
        else:
            !mv $curPath /content/traffic_signs/images/train
    else:
        if file.endswith(".txt"):
            !mv $curPath /content/traffic_signs/labels/validation/
        else:
            !mv $curPath /content/traffic_signs/images/validation/
print("Done")
```
